# Programas en red

El protocolo más usado es el protocolo web HTTP. Python tiene una biblioteca especial diseñada específicamente para soportar el protocolo HTTP para la recuperación de documentos y datos en la web, llamada **socket**.

Un socket es una conexión de doble sentido entre dos programas, en el cual se puede tanto leer como escribir en un mismo socket. Si se escribe algo en un socket, es enviado a la aplicación que está al otro lado de éste. Si se lee desde el socket, se obtienen los datos que la otra aplicación ha enviado.
 
El ejemplo siguiente muestra cómo hacer una conexión de red de bajo nivel con sockets.
Los sockets se pueden usar para comunicarse con un servidor web o con un servidor de correo o muchos otros tipos de servidores. Todo lo que se necesita es encontrar el documento y escribir el código para enviar y recibir los datos de acuerdo al protocolo.

Uno de los requisitos para usar el protocolo HTTP es la necesidad de enviar y recibir datos como objetos tipo "bytes", en lugar de cadenas (string). Los métodos .encode() y .decode() convierten cadenas en objetos tipo bytes y viceversa.

```python
import socket

# Creando un objeto object
mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Conectando al servidor
server_address = ('www.lamolina.edu.pe', 80)
mysock.connect(server_address)

#Enviando requerimiento GET
request = b'GET / HTTP/1.1\r\nHost: www.lamolina.edu.pe\r\n\r\n'
mysock.send(request)

# Recibiendo e imprimiendo la respuesta con paquetes de 512 bytes
while True:
    data = mysock.recv(512)
    if len(data) < 1:
        break
    print(data.decode(),end='')
    print(type(data))

# Cerrando el socket
mysock.close()
```

## Recuperando una imagen desde HTTP

Desde Python se puede recuperar una imagen a través del uso de HTTP. Se puede utilizar un programa similar que el codificado para leer un archivo de texto plano, pero en vez de copiar los datos en la pantalla cuando corre el programa, se acumularán los datos en un string, se recorta los encabezado y luego se guarda la imagen, como muestra el siguiente ejemplo:

```python
import socket
import time

HOST = 'data.pr4e.org'
PORT = 80
mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
mysock.connect((HOST, PORT))
mysock.sendall(b"GET http://data.pr4e.org/cover3.jpg HTTP/1.0\r\n\r\n")
count = 0
picture = b""

while True:
    data = mysock.recv(5120)
    if len(data) < 1: break
    #time.sleep(0.25) función que relantiza el tiempo entre llamada
    count = count + len(data)
    #print(len(data), count)
    picture = picture + data    
mysock.close()

#print(type(picture))
#print(picture)

# Buscando el fin del encabezado
pos = picture.find(b"\r\n\r\n")
print('Longitud del encabezado', pos)
print(picture[:pos+4].decode())


# Saltando el encabezado y guardando los datos de la imagen
picture = picture[pos+4:]
fhand = open("imagen.jpg", "wb")
fhand.write(picture)
fhand.close()
#Referencia: https://es.py4e.com/html3/12-network
```

## Recuperando páginas web con urllib

Si bien podemos enviar y recibir datos manualmente a través de HTTP utilizando la biblioteca de sockets, una manera mucho más simple de realizar esta tarea común en Python es utilizando la librería "urllib", la cual contiene varios módulos para trabajar con url.
Usando "urllib" se puede tratar una página web como un archivo. Simplemente se indica qué página web se desea recuperar y "urllib" maneja todos lo referido al protocolo HTTP protocolo y detalles del encabezado.
El código equivalente para leer el archivo romeo.txt de la web usando urllib es como sigue:

```python
import urllib.request

fhand = urllib.request.urlopen('http://www.lamolina.edu.pe')
#print(fhand)
for line in fhand:
    print(type(line))
    print(line.decode().strip())
```

Una vez que la página web se ha abierto con urllib.urlopen, podemos tratarla como un archivo y leerlo usando un bucle for. Cuando se ejecuta el programa, solo vemos la salida del contenido del archivo, los encabezados todavía se envían, pero el código urllib maneja los encabezados y solo devuelve los datos.

El siguiente ejemplo recuperar los datos de "romeo.txt" y calcule la frecuencia de cada palabra en el archivo y lo almacena en un dictionary

```python
import urllib.request, urllib.parse, urllib.error

fhand = urllib.request.urlopen('http://data.pr4e.org/romeo.txt')

counts = dict()   #{key1:content1,key2:content2 }
texto=""
for line in fhand:
    words = line.decode().split() 
    print(words)
    for word in words:
        counts[word] = counts.get(word, 0) + 1
print(counts)
```

## Leyendo archivos binarios usando urllib

En ocasiones se desea recuperar un archivo que no sea de texto (o binario), como una imagen o archivo de vídeo. Los datos en este tipo de archivos generalmente no son útiles para imprimir, pero se puede hacer una copia de una URL a un archivo local en el disco duro usando "urllib".
El procedimiento es abrir la URL a usar para descargar todo el contenido del documento en una variable de cadena (img) y luego escribir esa información en un archivo local de la siguiente manera:

```python
import urllib.request, urllib.parse, urllib.error

img = urllib.request.urlopen('http://data.pr4e.org/cover3.jpg').read()
#print(img)

fhand = open('graficoXYZ.jpg', 'wb')
fhand.write(img)
fhand.close()

#print(type(img))
```

Este programa funcionará si el tamaño del archivo es menor que el tamaño de la memoria de la computadora.
Sin embargo, si este es un archivo grande de audio o video, este programa puede fallar o al menos correr lentamente cuando la computadora se queda sin memoria. Para evitar quedarse sin memoria, se recupera los datos en bloques (o buffers) y luego se escribe cada bloque en el disco antes de recuperar el siguiente bloque. De esta manera el programa puede leer archivos de cualquier tamaño sin usar toda la memoria que tiene la computadora.

```python
import urllib.request, urllib.parse, urllib.error

img = urllib.request.urlopen('http://data.pr4e.org/cover3.jpg')
fhand = open('graphY.jpg', 'wb')
size = 0
while True:
    info = img.read(100000)
    if len(info) < 1: break
    print(len(info))    
    size = size + len(info)
    fhand.write(info)
    #print(size)

print(size, 'characters copied.')
fhand.close()
```
