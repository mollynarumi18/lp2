## Analizando HTML usando Expresiones Regulares

```python
import urllib.request, urllib.parse, urllib.error
import re

fhand = urllib.request.urlopen('https://www.octoparse.es/blog/30-mejores-software-gratuitos-de-web-scraping')
page=""


for i in fhand:
    page = page + i.decode().strip()
print(page)
    


#nota: en Anaconda prompt:     jupyter notebook --NotebookApp.iopub_data_rate_limit=1.0e10
```

```python
print(page)
```

```python
links = re.findall('href="(http[s]?://.*?)"', page)
for link in links:
    print(link)
```

## Analizando HTML usando BeautifulSoup

https://www.crummy.com/software/BeautifulSoup/bs4/doc/


```python
#En Anaconda prompt:   conda install -c anaconda beautifulsoup4 
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup

url = input('Entrar la url: ')    
#url=""
html = urllib.request.urlopen(url).read()
#print(type(html))
#print(type(html.decode()))
#print(len(html.decode()))

soup = BeautifulSoup(html, 'html.parser')
#print(type(soup))
print(soup.prettify())
```

```python
print(soup.find("title").text) # halla el título tag(o etiquetas) de la página web
```

```python
import re
#Sin beatyfulsoup utilizando solo expresiones regulares:
page=html.decode()
#print(page)
#print(page)
#print(type(page))
print(re.findall("<title>(.*)</title>",page))
```

```python
print(soup.find_all("p"))# halla todos las etiquetas p (párrafos)
print(len(soup.find_all("p"))) # cuenta el número de etiquetas p
print(type(soup.find_all("p"))) 
print(soup.find_all('p')[0].text)# cuenta el número de etiquetas p
```

```python
#print(len(soup.find_all('ul',"dropdown-menu")))
print(type((soup.find_all('ul',"dropdown-menu"))[2].text))
```

```python
print(soup.find('h2').text)
print(soup.find_all('h2'))
print(len(soup.find_all('h2')))
print(soup.find_all('h2')[0])
print(soup.find_all('h2')[1].text)
```

```python
#print(soup.find_all('a'))
print(type(soup.find_all('a')))
#print(soup.find('a'))
```

```python
L=[]
for heading in soup.find_all(["h1", "h2", "h3"]): # Recorre y busca todos las etiquetas con: h1, h2, h3
    L.append(heading)    #Almacena cada etiqueta hallada en una lista L
print(L)
```

```python
etiquetas = soup('a')
#print(type(etiquetas))
#print(len(etiquetas))
for i in etiquetas:
   #print(i)
    print(i.get('href', None))
```

```python
html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>

"""

soup2 = BeautifulSoup(html_doc, 'html.parser')

print(soup2.prettify())
```
