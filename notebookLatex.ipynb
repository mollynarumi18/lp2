## Raíces y Logaritmos
 
Puedes expresar una raíz cuadrada en LAtex usando `\sqrt` y para cambiar el nivel de la raíz puedes usar `\sqrt[n]` donde `n` es el nivel de la raíz.
 
$$f(x) = \sqrt[3]{2x} + \sqrt{x-2}$$
 
Para representar log usar `\log_{base}` donde `base` es la base del término del logaritmo.
 
$$\log_{x} x = 1$$

 
**Código:**

```markdown
$$f(x) = \sqrt[3]{2x} + \sqrt{x-2}$$
$$\log_{x} x = 1 $$
```
## Sumas & Productos
 
Puedes representar una suma con `\sum\limits_{a}^{b}`donde a y b son los límites de la suma.
 
$$\sum\limits_{x=1}^{\infty} {1 \over x} = 2$$
 
Puedes representar un producto con `\prod\limits_{a}^{b}` donde a y b son los límites del producto.
 
$$\prod\limits_{i=1}^{n} x_i - 1$$
 
**Código:**
```
$$\sum\limits_{x=1}^{\infty} {1 \over x} = 2$$
$$\prod\limits_{i=1}^{n} x_i - 1$$
```
